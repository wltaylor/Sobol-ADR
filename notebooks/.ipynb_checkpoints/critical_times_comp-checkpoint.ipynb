How much of a difference in critical times is there if we measure the times based on:
- Percentiles of mass passed through the system
- Percentiles of the peak concentration

```python
times = np.linspace(0,20,5000)

L = 2
x = 2
ts = 0.25
v = 1
Co = 1
theta = 0.7
rho_b = 1.5
dispersivity = 1
lamb = 0
alpha = 0.01
kd = 0.01

early_idx, peak_idx, late_idx, concentrations, adaptive_times = model.concentration_106_all_metrics_adaptive(times,theta,rho_b,dispersivity,lamb,alpha,kd, Co=Co, v=v, ts=ts, L=L, x=x)
early, peak, late = model.calculate_metrics(adaptive_times, concentrations)

print(f'Early: {early}')
print(f'Peak: {peak}')
print(f'Late: {late}')
```

```python
from scipy.integrate import trapz

def breakthrough(x_values, y_values, percentile):
    """
    Computes the time at which a given percentile of mass has passed
    Args:
        x_values: time values
        y_values: concentration values
        percentile: value between 0 and 1 representing desired percent of mass passed through system
    Returns:
        float: Time at which percentile of mass has passed
    """
    total_area = trapz(y_values, x_values) # calculates total area under the curve using trapezoidal rule
    cumulative_area = 0.0 # start point

    for i in range(1,len(x_values)):
        # incrementally sum BTC via trap rule
        dx = x_values[i] - x_values[i-1]
        avg_height = 0.5 * (y_values[i] + y_values[i-1])
        area_segment = dx * avg_height
        cumulative_area += area_segment

        if cumulative_area >= percentile * total_area:
            time = x_values[i]
            return time,i
```

```python
plt.plot(adaptive_times, concentrations, c='black')
plt.xlabel('Time')
plt.ylabel('Concentration (Co/C)')

plt.scatter(adaptive_times[early_idx], concentrations[early_idx], marker='o', s=75, c='blue', zorder=2)
plt.scatter(adaptive_times[peak_idx], concentrations[peak_idx], marker='o', s=75, c='blue', zorder=2)
plt.scatter(adaptive_times[late_idx], concentrations[late_idx], marker='o', s=75, c='blue', zorder=2)

time, index = breakthrough(adaptive_times, concentrations, 0.1)
plt.scatter(adaptive_times[index], concentrations[index], marker='o', s=75, c='red', zorder=2)
time, index = breakthrough(adaptive_times, concentrations, 0.9)
plt.scatter(adaptive_times[index], concentrations[index], marker='o', s=75, c='red', zorder=2)

plt.tight_layout()
plt.show()
```
